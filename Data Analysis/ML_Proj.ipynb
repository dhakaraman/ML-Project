```python
#importing all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Amazon

Raw Data

```python
#Analysing mean, standard deviation, max and min value for each column
df_amazon = pd.read_csv('AMZN.csv') #reading the csv file
print(df_amazon.info()) #printing the metadata 
index = range(1, len(df_amazon) + 1) 
df_amazon.insert(0, 'Index', index) 
df_amazon.pop('Adj Close')
print(df_amazon.head())
print('\n')
print("Open")
print("Mean= ", np.mean(df_amazon['Open']),  "  \t Std= ", np.std(df_amazon['Open']),  "  \t Max= ", np.max(df_amazon['Open']),  "  \t Min= ", np.min(df_amazon['Open']))
print('High')
print("Mean= ", np.mean(df_amazon['High']),  "  \t Std= ", np.std(df_amazon['High']),  "  \t Max= ", np.max(df_amazon['High']),  "  \t Min= ", np.min(df_amazon['High']))
print('Low')
print("Mean= ", np.mean(df_amazon['Low']),   "  \t Std= ", np.std(df_amazon['Low']),   "  \t Max= ", np.max(df_amazon['Low']),   "  \t Min= ", np.min(df_amazon['Low']))
print('Close')
print("Mean= ", np.mean(df_amazon['Close']), "  \t Std= ", np.std(df_amazon['Close']), "  \t Max= ", np.max(df_amazon['Close']), "  \t Min= ", np.min(df_amazon['Close']))
print('Volume')
print("Mean= ", np.mean(df_amazon['Volume']),"  \t Std= ", np.std(df_amazon['Volume']),"  \t Max= ", np.max(df_amazon['Volume']),"  \t Min= ", np.min(df_amazon['Volume']))
```


Data visualisation

```python
#Plotting open,high,low and close value to visualize the data and its trend
plt.xlabel('Index')
plt.ylabel('Price(USD)')
plt.plot(df_amazon['Index'],df_amazon['Open'],label='Open')
plt.plot(df_amazon['Index'],df_amazon['High'],label='High')
plt.plot(df_amazon['Index'],df_amazon['Low'],label='Low')
plt.plot(df_amazon['Index'],df_amazon['Close'],label='Close')
plt.legend()
plt.show()
```

Index vs Volume

```python
#plot of volume feature
plt.xlabel('Index')
plt.ylabel('Volume')
plt.style.use('seaborn')
plt.plot(df_amazon['Index'],df_amazon['Volume'])
plt.show()
```

Normalized data

```python
#Normalising data using minmax scaler and we have scaled data between 0-1
from sklearn.preprocessing import MinMaxScaler
def normalise(df):
    #https://machinelearningmastery.com/standardscaler-and-minmaxscaler-transforms-in-python/
    scaler = MinMaxScaler()
    numeric = ['Open','High','Low', 'Close', 'Volume'] #normalising the columns which are numeric in nature
    df[numeric] = scaler.fit_transform(df[numeric])

    return df
```

```python
df_amazon = normalise(df_amazon)
df_amazon.to_csv('amazon.csv', index=False)
```

```python
#Analysing mean, standard deviation, max and min value for each column
print(df_amazon.head())
print('\n')
print("Open")
print("Mean= ", np.mean(df_amazon['Open']),  "  \t Std= ", np.std(df_amazon['Open']),  "  \t Max= ", np.max(df_amazon['Open']),  "  \t Min= ", np.min(df_amazon['Open']))
print('High')
print("Mean= ", np.mean(df_amazon['High']),  "  \t Std= ", np.std(df_amazon['High']),  "  \t Max= ", np.max(df_amazon['High']),  "  \t Min= ", np.min(df_amazon['High']))
print('Low')
print("Mean= ", np.mean(df_amazon['Low']),   "  \t Std= ", np.std(df_amazon['Low']),   "  \t Max= ", np.max(df_amazon['Low']),   "  \t Min= ", np.min(df_amazon['Low']))
print('Close')
print("Mean= ", np.mean(df_amazon['Close']), "  \t Std= ", np.std(df_amazon['Close']), "  \t Max= ", np.max(df_amazon['Close']), "  \t Min= ", np.min(df_amazon['Close']))
print('Volume')
print("Mean= ", np.mean(df_amazon['Volume']),"  \t Std= ", np.std(df_amazon['Volume']),"  \t Max= ", np.max(df_amazon['Volume']),"  \t Min= ", np.min(df_amazon['Volume']))
```

```python
#Plotting open,high,low and close value to visualize the data and its trend
plt.xlabel('Index')
plt.ylabel('Price')
plt.plot(df_amazon['Index'],df_amazon['Open'],label='Open')
plt.plot(df_amazon['Index'],df_amazon['High'],label='High')
plt.plot(df_amazon['Index'],df_amazon['Low'],label='Low')
plt.plot(df_amazon['Index'],df_amazon['Close'],label='Close')

plt.legend()
plt.show()
```
